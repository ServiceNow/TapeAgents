```python
import pandas as pd
```

```python
df = pd.read_csv("error_analysis.csv")
# set nan in the comment column to empty string
df["comment"] = df["comment"].fillna("")
# set nan values to 0
df.fillna(0, inplace=True)
df.head()
```

```python
import sys

sys.path.append("../../..")
from examples.gaia_agent.eval import tape_correct
from examples.gaia_agent.tape import GaiaTape
from tapeagents.io import load_tapes

gpt4o_tapes_path = "../../../../gaia/runs/gpt4o_t02_1/tapes"
tapes = load_tapes(GaiaTape, gpt4o_tapes_path, file_extension=".json")
for tape in tapes:
    print(tape.metadata.task["task_id"], int(tape_correct(tape)))
    # add the correct score to the dataframe
    df.loc[df["task_id"] == tape.metadata.task["task_id"], "gpt4o_correct"] = int(tape_correct(tape))
```

```python
df.head()
```

```python
# print solved percent by level
for level in range(1, 4):
    level_df = df[df["level"] == level]
    print(f"Level {level}: {level_df['solved'].mean() * 100:.2f}% solved")
    print(f"Level {level}: {level_df['gpt4o_correct'].mean() * 100:.2f}% solved by GPT-4O")
    print(f"Level {level}: {level_df['solved'].sum()} out of {level_df.shape[0]} solved")
    print(f"Level {level}: {level_df['gpt4o_correct'].sum()} out of {level_df.shape[0]} solved by GPT-4O")
# total solved percent
print(f"Total: {df['solved'].mean() * 100:.2f}% solved")
print(f"Total: {df['gpt4o_correct'].mean() * 100:.2f}% solved by GPT-4O")
```

```python
# print solved by gpt4o per error type
error_columns = [
    "bad plan",
    "bad doc parsing",
    "bad image reading",
    "cannot read video",
    "cannot find info",
    "read page wrong",
    "failed to follow plan",
    "bad reasoning",
    "browser errors",
    "wrong output format",
    "not enough steps",
]
for error_column in error_columns:
    error_df = df[df[error_column] == 1]
    print(f"{error_column}: {error_df['gpt4o_correct'].mean() * 100:.2f}% solved by GPT-4O")
    print(f"{error_column}: {error_df['gpt4o_correct'].sum()} out of {error_df.shape[0]} solved by GPT-4O")
```
